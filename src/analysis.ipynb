```python
from pysus import SINAN
import pandas as pd
```

# Data acquisition

```python
sinan = SINAN().load()
```

```python
files = sinan.get_files(dis_code=['MALA'])
```

```python
parquet = sinan.download(files)
```

```python
df = None
for p in parquet:
    p.to_dataframe()
    df = pd.concat([df, p.to_dataframe()])
```

```python
display(df)
```

```python
df.to_csv('malaria.csv', index=False)
```
